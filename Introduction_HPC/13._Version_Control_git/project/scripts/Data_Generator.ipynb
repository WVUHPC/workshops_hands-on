```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(0.1,2*np.pi,0.01)
```

```python
Y=np.sin(X)+np.cos(10*X)
```

```python
plt.plot(X,Y,'.')
fig1 = plt.gcf()
fig1.savefig('../plots/wave.pdf')
```

```python
data=np.stack((X,Y)).T
```

```python
np.savetxt('../data/wave.dat', data)
```
